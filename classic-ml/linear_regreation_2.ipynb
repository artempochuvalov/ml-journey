```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X = np.array([
	[30, 1, 1],
	[40, 2, 2],
	[50, 3, 1],
	[60, 4, 3],
	[80, 5, 2],
	[100, 6, 3],
])
y = np.array([
	120000,
	150000,
	170000,
	200000,
	240000,
	280000
])
```

```python
X_aug = np.hstack([np.ones((X.shape[0], 1)), X])
```

```python
w = (np.linalg.inv(X_aug.T @ X_aug) @ X_aug.T @ y)
model = LinearRegression().fit(X, y)
print(f'Веса, найденные по формуле: {w}')
print(f'Веса, найденные с помощью sklearn: {w}')
```

```python
property_names = ['Площадь', 'Этаж', 'Район']

for i in range(X.shape[1]):
	plt.figure()
	plt.scatter(X[:, i], y, color='blue', label=f'{property_names[i]}/цена')
	plt.plot(X[:, i], X_aug @ w, color='red', label='Линейная регрессия, вычисленная по формуле', linewidth=5)
	plt.plot(X[:, i], model.predict(X), color='orange', label='Линейная регрессия, вычисленная с помощью sklearn')
	plt.xlabel(property_names[i])
	plt.ylabel('Цена')
	plt.legend()
	plt.show()
```
