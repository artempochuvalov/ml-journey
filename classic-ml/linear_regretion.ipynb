```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
X = np.array([
	[30],
	[40],
	[50],
	[60],
	[80],
	[100],
])
y = np.array([
	120000,
	150000,
	170000,
	200000,
	240000,
	280000
])
X_aug = np.hstack([np.ones((X.shape[0], 1)), X])
```

```python
w = np.linalg.inv(X_aug.T @ X_aug) @ (X_aug.T @ y)
```

```python
print(f'Weights of linear regration calculated my self:\n{w[1:]}')
```

```python
model = LinearRegression(copy_X=True).fit(X, y)
```

```python
print(f'Weights of linear regration calculated by sklearn:\n{model.coef_}')
```

```python
plt.scatter(X, y, color='blue', label='Данные')
plt.plot(X, model.predict(X), color='red', label='Линейная регрессия')
plt.xlabel('Площадь')
plt.ylabel('Цена')
plt.legend()
plt.show()
```

Предсказание данных

```python
new_X = np.array([[70], [90]])
predicted = model.predict(new_X)
print(f'Predicted by sklearn: {predicted}')
```
